```python
#Create new columns from the nested dictionary:

df = df.copy()

df["sale_price"] = df["prices"].apply(
    lambda x: x.get("price") if isinstance(x, dict) else None
)

df["regular_price"] = df["prices"].apply(
    lambda x: x.get("regular_price") if isinstance(x, dict) else None
)
```

```python
#Now select useful columns:
selected_cols = [
    "id",
    "name",
    "sale_price",
    "regular_price"
]

df = df[selected_cols].copy()
```

```python
#Rename columns:

df.columns = [
    "product_id",
    "product_name",
    "sale_price",
    "regular_price"
]
```

```python
#Feature Engineering
#Convert prices to numeric:

import pandas as pd

df["sale_price"] = pd.to_numeric(
    df["sale_price"],
    errors="coerce"
)

df["regular_price"] = pd.to_numeric(
    df["regular_price"],
    errors="coerce"
)
```

```python
#Create Discount:

df["discount"] = (
    df["regular_price"]
    - df["sale_price"]
)
```

```python
#Create Discount Percentage:

df["discount_percentage"] = (
    df["discount"]
    / df["regular_price"]
) * 100
```

```python
#Save Clean Data
#First create the folder if it doesn't exist:
import os

os.makedirs("output", exist_ok=True)
```

```python
 # Then save
df.to_csv(
    "output/clean_products.csv",
    index=False
)
```
